```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


# Captura desde la webcam
cam = cv2.VideoCapture(0)
ret, frame = cam.read()
cam.release()

if ret:
    # Guardar la imagen original
    cv2.imwrite("foto_original.jpg", frame)
    print("✅ Imagen capturada")
else:
    print("❌ No se pudo capturar imagen")
```

```python
# Leer en formato RGB para matplotlib
img_original = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

plt.imshow(img_original)
plt.title("Imagen Original")
plt.axis("off")
plt.show()
```

```python
img_invertida = 255 - img_original

plt.imshow(img_invertida)
plt.title("Imagen Invertida")
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
_, bw = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(bw, cmap="gray")
plt.title("⚫ Imagen en blanco y negro")
plt.axis("off")
plt.show()
```

```python
img_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap='gray')
plt.title("Imagen en Escala de Grises")
plt.axis("off")
plt.show()
```

```python

img_roja = img_original.copy()
img_roja[:, :, 0] = 0  # Quitar azul
img_roja[:, :, 1] = 0  # Quitar verde

plt.imshow(img_roja)
plt.title("Imagen Roja Intensa")
plt.axis("off")
plt.show()
```

```python
img_azul = img_original.copy()
img_azul[:, :, 1] = 0  # Quitar verde
img_azul[:, :, 2] = 0  # Quitar rojo

plt.imshow(img_azul)
plt.title("Imagen Azul Intensa")
plt.axis("off")
plt.show()
```

```python
# Convertir de RGB a BGR antes de guardar
cv2.imwrite("imagen_roja.jpg", cv2.cvtColor(img_roja, cv2.COLOR_RGB2BGR))
cv2.imwrite("imagen_azul.jpg", cv2.cvtColor(img_azul, cv2.COLOR_RGB2BGR))

print("✅ Imágenes guardadas como 'imagen_roja.jpg' y 'imagen_azul.jpg'")
```
